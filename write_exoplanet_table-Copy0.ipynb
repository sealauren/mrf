```python
#%matplotlib osx
%pylab inline
import numpy as np
import math
#import astroypy
import pandas as pd
import matplotlib.pyplot as plt
dtype_dict = {'NAME': str, 'MSINI': np.float32, 'MSINIUPPER': np.float32, 'MSINILOWER': np.float32,
              'UMSINI':np.float32, 'R':np.float32, 'RUPPER':np.float32, 'RLOWER':np.float32, 'UR':np.float32,
              'MSTAR':np.float32, 'MSTARUPPER':np.float32, 'MSTARLOWER':np.float32, 'UMSTAR':np.float32, 
              'RSTAR':np.float32, 'RSTARUPPER':np.float32, 'RSTARLOWER':np.float32, 'URSTAR':np.float32,
              'TEFF':np.float32, 'TEFFUPPER':np.float32, 'TEFFLOWER':np.float32, 'UTEFF':np.float32,
              'A':np.float32, 'AUPPER':np.float32, 'ALOWER':np.float32, 'UA':np.float32,
              'PER':np.float32, 'ECC':np.float32, 'I':np.float32,
              'FIRSTREF':str, 'FIRSTURL':str, 'ORBREF':str, 'ORBURL':str, 'DATE':int}
df = pd.read_csv("all_exos_errors6.csv",skiprows=[1])
df.sort(inplace=True)
df.head()
```

```python
new_columns = [c.lower() for c in df.columns]
df.columns = new_columns
print df.columns
```

```python
sini = df.i*np.pi/180.
df['m'] = df.msini/sini
df['um'] = df.umsini
df.columns
```

```python
# Need to build local lookup table, or use wolfram_lookup, to get common values
#%load 'astrophysical_constants.txt'
```

```python
mearth = 5.9721986E27 #grams
rearth = 6.3675E8 #centimeters
msun = 1.988435E33 #grams
rsun = 6.955E10 #centimeters
cmperau=1.496E13
lsun = 3.846E33 #ergs per second
tsun = 5780 #kelvins
```

```python
df['rho']= df.m*mearth/(4/3*np.pi*(df.r*rearth)**3) # cgs
df['urho'] = np.sqrt((df.umsini/df.m)**2. + (3*df.ur/df.r)**2.)*df.rho
```

```python
df22 = pd.read_csv('h22_orbital_pars_annotated.txt',skiprows=[1],skipinitialspace=True)
df22_stars = pd.read_csv('h22_star_prop.txt',skipinitialspace=True)
```

```python
df22.columns = [c.strip() for c in df22.columns]
```

```python
new_columns = [c.lower() for c in df22.columns]
df22.columns = new_columns
df22.columns = ['name','per','r','ur','m','um','rho','urho','k','uk','rhostar','urhostar','b','ub','rprs','tt','mupper']
```

```python
df22_stars.koi_star = [koi.strip().lstrip('k0') for koi in df22_stars.koi_star]
df22_stars
```

```python
# look up how to group stars with planets: use join method?
df22['koi_star']=[k.strip()[:-3] for k in df22.name]
df22_full = pd.merge(df22,df22_stars)
df22_full
```

Use Kepler's Law to get semi-major axis, flux for H22 planets.

```python
df22_full['a'] = pow(pow(df22_full.per/365.25,2.) / df22_full.mstar , 3.) * cmperau
df = df.append(df22_full)
```

```python
flux_earth = lsun/(4*np.pi*cmperau**2.) # flux on earth
df['flux']=df.rstar**2*(df.teff/5778.)**4/df.a**2 #* np.sqrt(1./(1 - df.ecc**2))  #time avg. flux on the planet in earth fluxes
df['uflux'] = df.flux * np.sqrt((4*df.uteff/df.teff)**2. + (2*df.urstar/df.rstar)**2 + (2*df.ua/df.a)**2)
medf = np.median(df.flux)
df_hot = df[df.flux > medf]
df_cold = df[df.flux <= medf]
```

```python
fig = plt.figure(figsize=(12,12))
ax1 = fig.add_subplot(2,1,1, xscale="log", yscale="log") # I don't understand what's going on here.
ax2 = fig.add_subplot(2,1,2, xscale="log", yscale="log") # I don't understand what's going on here.
#ax1.errorbar(df.m, df.r, xerr=df.umsini, yerr=df.ur, fmt='o', markersize=5, 
#             alpha =0.5, color='k'); # black option
ax1.errorbar(df_hot.m, df_hot.r, xerr=df_hot.um, yerr=df_hot.ur, 
             fmt='o', markersize=5, 
             alpha=0.5, color='r', label='$F > $ %i $F_E$'%medf, 
             capsize=0);
ax1.errorbar(df_cold.m, df_cold.r, xerr=df_cold.um, yerr=df_cold.ur, 
             fmt='o', markersize=5, 
             alpha =0.5, color='b', label='$F < $ %i $F_E$'%medf, 
             capsize=0);
ax2.errorbar(df_hot.m, df_hot.rho, xerr=df_hot.um, yerr=df_hot.urho, 
             fmt='o', markersize=5, 
             alpha=0.5, color='r', label='$F > $ %i $F_E$'%medf, 
             capsize=0);
ax2.errorbar(df_cold.m, df_cold.rho, xerr=df_cold.um, yerr=df_cold.urho, 
             fmt='o', markersize=5, 
             alpha=0.5, color='b', label='$F < $ %i $F_E$'%medf,
             capsize=0);

fig.suptitle('Exoplanets', fontsize=22)
ax2.set_xlabel('Mass ($M_E$)', fontsize=16)
ax1.set_ylabel('Radius ($R_E$)',fontsize=16)
ax1.legend(loc='lower right');
ax2.set_ylabel('Density (g/cc)',fontsize=16)
ax2.legend(loc='lower right');
```
