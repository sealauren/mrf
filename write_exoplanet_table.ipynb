```python
sp = df[df.r <= 4.]
sp.loc[:,['m','r','orbref']]
```

```python
fitfunc = lambda p, x: p[0] + p[1] * x
errfunc = lambda p, x, y, err: ((y - fitfunc(p, x)) / err)**2.   
from scipy.optimize import leastsq
out = leastsq(errfunc,  [0.2,2.5], args=(sp.r, sp.m, sp.um),full_output=1)
print "Best parameters:", out[0]
print "Covariance:", out[1]
print "Chi^2:", errfunc(out[0],sp.r,sp.m,sp.um).sum()/(len(sp.m) - len(out[0]))
best_pars = out[0]
rarr = linspace(0.0, 4.0, 11)
print rarr
```

```python
rw, urw, mw, umw, rhow, urhow = [], [], [], [], [], []
for r in xrange(0,4):
    rbin = sp.loc[r <= sp.r].loc[sp.r < r+1]
    rw.append(np.average(rbin['r'], weights = (1./rbin.um**2. + 1./rbin.ur**2.)))
    mw.append(np.average(rbin['m'], weights = 1./rbin.um**2.))
    rhow.append(np.average(rbin['rho'], weights = 1./rbin.urho**2.))
    urw.append(std(rbin.r)/sqrt(len(rbin)))
    umw.append(std(rbin.m)/sqrt(len(rbin)))
    urhow.append(std(rbin.rho)/sqrt(len(rbin)))
print rw
print mw
print rhow
```

Calculate "residual" mass / radius for each planet:

```python
rpred = (sp.m - best_pars[0])/best_pars[1]
mpred = fitfunc(best_pars,sp.r)
r_resid = sp.r - rpred
m_resid = sp.m - mpred
```
