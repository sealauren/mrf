Load planets from Geoff's paper and merge the planetary and stellar tables.

```python
df22 = pd.read_csv('h22_orbital_pars_annotated.txt',skiprows=[1],skipinitialspace=True)
df22_stars = pd.read_csv('h22_star_prop.txt',skipinitialspace=True)
df22.columns = [c.strip() for c in df22.columns]
new_columns = [c.lower() for c in df22.columns]
df22.columns = new_columns
df22.columns = ['name','per','r','ur','m','um','rho','urho','k','uk','rhostar','urhostar','b','ub','rprs','tt','mupper']
df22_stars.koi_star = [koi.strip().lstrip('k0') for koi in df22_stars.koi_star]
df22_stars
df22_stars['fe'] = df22_stars.feh
df22_stars['ufe'] = df22_stars.ufeh
df22['koi_star']=[k.strip()[:-3] for k in df22.name]
df22_full = pd.merge(df22,df22_stars)
df22_full['name'] = ["KOI-"+koi for koi in df22_full.name]
df22_full.head()
```

Use Kepler's Law to get semi-major axis H22 planets and merge.

```python
df22_full['a'] = pow(pow(df22_full.per/365.25,2.) * df22_full.mstar , 1/3.)
df22_full['ecc'] = [0.0] * len(df22_full)
df22_full['orbref'] = "Marcy 2013"
df22_full['firstref'] = "Borucki 2011"
df = df.append(df22_full)
```

```python
# propagate Ms and Rs to new densities
df.loc[['55 Cnc e','CoRoT-7 b','GJ 1214 b'],'rho'] = \
df.loc[['55 Cnc e','CoRoT-7 b','GJ 1214 b'],'m']/(4/3. * pi* (df.loc[['55 Cnc e','CoRoT-7 b','GJ 1214 b'],'r'])**3.)*mearth/rearth**3.
df.loc[['55 Cnc e','CoRoT-7 b','GJ 1214 b'],'urho'] = df.loc[['55 Cnc e','CoRoT-7 b','GJ 1214 b'],'rho'] * \
np.sqrt((3 * df.loc[['55 Cnc e','CoRoT-7 b','GJ 1214 b'],'ur']/df.loc[['55 Cnc e','CoRoT-7 b','GJ 1214 b'],'r'])**2. + \
        (df.loc[['55 Cnc e','CoRoT-7 b','GJ 1214 b'],'um']/df.loc[['55 Cnc e','CoRoT-7 b','GJ 1214 b'],'m'])**2.)
df.loc[['55 Cnc e','CoRoT-7 b','GJ 1214 b'],['rho','urho']]
```

```python
flux_earth = lsun/(4*np.pi*cmperau**2.) # flux on earth
df['flux']=df.rstar**2*(df.teff/5778.)**4/df.a**2 * np.sqrt(1./(1 - df.ecc**2))  #time avg. flux on the planet in earth fluxes
df['uflux'] = df.flux * np.sqrt((4*df.uteff/df.teff)**2. + (2*df.urstar/df.rstar)**2 + (2*df.ua/df.a)**2)
medf = np.median(df.flux)
df_hot = df[df.flux > medf]
df_cold = df[df.flux <= medf]
```
